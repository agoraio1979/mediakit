```python
# Engagement rate by reach (ERR)
import json

df['ERR']= (df['Commenti'] + df['Mi piace'] + df['Salvataggi'] + df['Condivisioni'])/df['Copertura'] * 100
average_ERR = df['ERR'].mean()
print(f'Average ERR: {average_ERR:.2f}% over {len(df)} posts')
with open('public/data.json', 'r') as f:
    data=json.load(f)

    
data['average_ERR']=f'{average_ERR:.2f}'
with open('public/data.json', 'w') as f:
    json.dump(data, f, indent=2) 
```

```python
# Engagement rate by post
total_follower=40600
df['ERP']= (df['Commenti'] + df['Mi piace'] + df['Salvataggi'] + df['Condivisioni'])/total_follower * 100
average_ERP = df['ERP'].mean()
print(f'Average ERP: {average_ERP:.2f}% over {len(df)} posts')

with open('public/data.json', 'r') as f:
    data=json.load(f)

    
data['average_ERP']=f'{average_ERP:.2f}'
with open('public/data.json', 'w') as f:
    json.dump(data, f, indent=2) 
```

```python
# Engagement rate by impressions

df['ERI']= (df['Commenti'] + df['Mi piace'] + df['Salvataggi'] + df['Condivisioni'])/df['Visualizzazioni'] * 100
average_ERI = df['ERI'].mean()
print(f'Average ERI: {average_ERI:.2f}% over {len(df)} posts')

with open('public/data.json', 'r') as f:
    data=json.load(f)

    
data['average_ERI']=f'{average_ERI:.2f}'
with open('public/data.json', 'w') as f:
    json.dump(data, f, indent=2) 
```

```python
import pandas as pd
from io import StringIO

with open('Pubblico.csv', 'r',encoding='utf-16') as f:
    blocks=f.read().split('\n\n')
    
    
dfs=[]
    
for block in blocks:
    dfs.append(pd.read_csv(StringIO(block)))
```

```python
with open('public/data.json', 'r') as f:
    data=json.load(f)

    
data['age']=[{row[1]['Unnamed: 0']} for row in dfs[1].iterrows()]
data['gender']=[{row[1]['Unnamed: 0']} for row in dfs[1].iterrows()]
data['locations']
with open('public/data.json', 'w') as f:
    json.dump(data, f, indent=2) 
```

## Followers

```python
import pandas as pd
import json
df=pd.read_csv('Follower.csv',encoding='utf-16')
df
followerAtStart=11125 # ~ al 12 gennaio 2025
df['totalFollowers']=df['Primary'].cumsum() + followerAtStart
months={
    1:'Gen',
    2:'Feb',
    3:'Mar',
    4:'Apr',
    5:'Mag',
    6:'Giu',
    7:'Lug',
    8:'Ago',
    9:'Set',
    10:'Ott',
    11:'Nov',
    12:'Dic'
}


df['labels']=pd.to_datetime(df['Data']).dt.day.astype(str) + ' ' + pd.to_datetime(df['Data']).dt.month.map(months).astype(str)    
df
```

```python
with open('public/data.json', 'r') as f:
    data=json.load(f)

    
data['followerGrowth']['labels']=df['labels'].tolist()
data['followerGrowth']['values']=df['totalFollowers'].tolist()
with open('public/data.json', 'w') as f:
    json.dump(data, f, indent=2) 
```
